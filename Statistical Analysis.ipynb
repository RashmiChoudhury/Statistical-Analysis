```python
import pandas as pd
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plot
playstore_data=pd.read_csv('googleplaystore.csv')
playstore_user_review=pd.read_csv('googleplaystore_user_reviews.csv')
```

```python
#the contents of Google playstore file

playstore_data
```

```python
#the contents of playstore user review file

playstore_user_review
```

```python
#Data cleaning

filtered_data= playstore_data[~playstore_data.Category.str.contains(r'\d')]
filtered_playstore_data =filtered_data.dropna()
filtered_playstore_data
```

```python
#Data cleaning to drop missing values in rows

filtered_reviews= playstore_user_review.dropna()
filtered_reviews
```

```python
#What are the different catagories of Apps and how many apps are in each Category?

grouped_playstore_data = filtered_playstore_data.groupby(["Category"]).count()
grouped_playstore_data[['App']]
```

```python
#What is the percentage of Apps under each Category ? 

total=grouped_playstore_data['App'].sum()
def calculate_percentage(App_count_cat):
    percentage_is = (App_count_cat/total) * 100
    return percentage_is

grouped_playstore_data['percentage_is'] = grouped_playstore_data.App.apply(calculate_percentage)
grouped_playstore_data[['App','percentage_is']]
```

```python
#Plot Bar graph to visualise percentage of Apps covered by each category

bar_graph= grouped_playstore_data[['percentage_is']]
bar_graph.plot.bar(figsize=(20,10))
```

```python
from bokeh.models.tools import HoverTool
from bokeh.sampledata.glucose import data
from bokeh.plotting import figure
from bokeh.io import output_notebook, show

x = filtered_playstore_data[['App']]
y = filtered_playstore_data[['Rating']]

# Basic plot setup
p = figure(width=900, height=300, x_axis_type="App", title='Hover over points')

p.line(x,y, line_dash="4 4", line_width=1, color='gray')

cr = p.circle(x,y, size=20,
              fill_color="yellow", hover_fill_color="firebrick",
              fill_alpha=0.05, hover_alpha=0.3,
              line_color=None, hover_line_color="white")

p.add_tools(HoverTool(tooltips=None, renderers=[cr], mode='hline'))

show(p)
```

```python
#Selecting top 10 categories in terms of number of Apps and adding up the remaining number of Apps to form 'Others' Category 

grouping=filtered_playstore_data[['Category','App']]
pie_chart=grouping.replace(grouping.groupby('Category').count().sort_values('App',ascending=False).index[10:],'Others').groupby('Category').count()
pie_chart
```

```python
#Plot pie chart to visualise top 10 Categories with most number of Apps and Others

pie_chart.plot.pie(y='App', figsize=(10, 10),autopct='%1.0f%%')
```

```python
'''Categorize the Apps according to their ratings for Types- Poor (if Rating < 2.5), Average (if 2.6<Rating<3.9) and Good 
(if Rating>4)'''
#Find the number of Apps in each Type

app_vs_rating= filtered_playstore_data[['App','Rating']]
app_vs_rating.loc[app_vs_rating.Rating < 2.5, 'Type' ] = 'Poor'
app_vs_rating.loc[(app_vs_rating.Rating >2.6) & (app_vs_rating['Rating'] <3.9) , 'Type'] = 'Average'  
app_vs_rating.loc[app_vs_rating.Rating >4 , 'Type'] = 'Good'

arranged_by_type=app_vs_rating
each_Type= app_vs_rating.groupby(['Type']).size()
each_Type
```
